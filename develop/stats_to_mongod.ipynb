Uploading the data from statscan through json. Needs to be uploaded with a string and then the first two elements cut off, after that jsonify.

```python
import pandas as pd
import json
import urllib.request

from flask import Flask, render_template, redirect,url_for, session
from flask_pymongo import PyMongo
import pymongo
```

```python
def getdata(url):
    #https://www12.statcan.gc.ca/wds-sdw/2016ind1-eng.cfm
    data = urllib.request.urlopen(url).read()

    #Statscan data has a 'head' that needs to be removed, then the json can be read properly.
    data1 = data[2:]
    output = json.loads(data1)
   # content = json.dumps(output, indent = 4, sort_keys=True)

    return output
```

```python
app = Flask(__name__)
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
db =    client.analysis
app.config["MONGO_URI"] = "mongodb://localhost:27017/analysis"
mongo = PyMongo(app)
```

```python
#This program rreads in the statscan data for one specific district. It is not complete, probably not needed, 
#because we will be going the *.csv route.
url = "https://www12.statcan.gc.ca/rest/census-recensement/CPR2016.json?lang=E&dguid=2016A000011124&topic=10&notes=0"
 
collection = mongo.db.canada
collection.update({}, getdata(url), upsert=True)
```
